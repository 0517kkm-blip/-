```
import pandas as pd
from scipy import stats
import os

zip_path = "/content/sample_data/type3/type3_sample_csvs (1).zip"
target_dir = "/content/sample_data/type3/"

if os.path.exists(zip_path):
    !unzip -o "{zip_path}" -d "{target_dir}"
    print("압축 해제가 완료되었습니다.")

file_path = "/content/sample_data/type3/branch_satisfaction.csv"
df = pd.read_csv(file_path)


a = df[df["branch"] == "A"]["score"]
b = df[df["branch"] == "B"]["score"]
c = df[df["branch"] == "C"]["score"]


result = stats.f_oneway(a, b, c)

means = df.groupby("branch")["score"].mean()


print("\n" + "="*30)
print("지점별 평균:")
print(means)
print("-" * 30)
print(f"F-statistic: {result.statistic:.4f}")
print(f"p-value: {result.pvalue:.4f}")
print("-" * 30)
print("평균이 가장 높은 지점:", means.idxmax())
print("="*30)
```

```
import pandas as pd
from scipy import stats
import os


zip_path = "/content/sample_data/type3/type3_sample_csvs (1).zip"
target_dir = "/content/sample_data/type3/"

if os.path.exists(zip_path):
    !unzip -o "{zip_path}" -d "{target_dir}"
    print("압축 해제가 완료되었습니다.")


file_path = "/content/sample_data/type3/before_after_worktime.csv"
df = pd.read_csv(file_path)


before = df["before_min"]
after = df["after_min"]
diff = before - after


result = stats.ttest_rel(before, after, alternative="greater")


print("\n" + "="*30)
print(f"평균 차이(before - after): {diff.mean():.4f}")
print(f"t-statistic: {result.statistic:.4f}")
print(f"p-value: {result.pvalue:.4f}")
print("="*30)
```
