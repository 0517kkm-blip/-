```
import pandas as pd
from scipy import stats
import numpy as np

# 1. 데이터 불러오기 (정확한 경로 사용)
df = pd.read_csv("/content/sample_data/delivery_time.csv")
x = df["delivery_min"].dropna()

# 2. 단일표본 t-test (대립가설: 평균 < 30)
result = stats.ttest_1samp(x, popmean=30, alternative="less")

# 3. 기초 통계량 및 신뢰구간 계산
mean_x = x.mean()
n = len(x)
se = stats.sem(x)
ci_low, ci_high = stats.t.interval(
    0.95, df=n-1, loc=mean_x, scale=se
)

# 4. 결과 출력
print("표본평균:", mean_x)
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
print("95% CI:", (ci_low, ci_high))
```

```
import pandas as pd
from scipy import stats

# 1. 데이터 불러오기 (type3 폴더 경로 추가)
df = pd.read_csv("/content/sample_data/type3/before_after_worktime.csv")

# 2. 변수 설정
before = df["before_min"]
after = df["after_min"]

# 3. 차이 계산 (before - after)
diff = before - after

# 4. 대응표본 t-test 수행 (대립가설: 교육 전 시간 > 교육 후 시간)
result = stats.ttest_rel(before, after, alternative="greater")

# 5. 결과 출력
print("평균 차이(before - after):", diff.mean())
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
```
